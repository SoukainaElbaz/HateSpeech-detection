```python
# Appliquer sur le dataset complet
df = pd.read_csv('../data/raw/HateSpeechData.csv')
df['clean_text'] = df['tweet'].apply(full_preprocess)

# Vérification
df[['tweet', 'clean_text']].head(10)
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# Textes vides après preprocessing
empty = df['clean_text'].apply(lambda x: x == '')
print(f'Textes vides après preprocessing : {empty.sum()} / {len(df)}')
df = df[~empty].reset_index(drop=True)

# Sauvegarde
df.to_csv('../data/processed/dataset_clean.csv', index=False)
print('Sauvegardé dans data/processed/dataset_clean.csv')
```
