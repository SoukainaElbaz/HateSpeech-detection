 # 1. text cleaning

```python
def clean_text(text, lang="en"):
    if not isinstance(text, str):
        text = str(text)
    if not text:
        return text

    # Emojis -> :alias: (before removing punctuation)
    text = emoji.demojize(text)
    text = text.lower()

    # Digits -> spelled-out words (e.g. 42 -> forty-two in English)
    def _digits_to_words(m):
        try:
            return num2words(int(m.group(0)), lang=lang)
        except (ValueError, OverflowError, NotImplementedError):
            return m.group(0)

    text = re.sub(r"\d+", _digits_to_words, text)
    # "forty-two" uses '-', which is not a word char — turn it into a space
    text = text.replace("-", " ")
    text = re.sub(r"[^\w\s]", "", text) # Removes punctuation and symbols
    text = re.sub(r"\s+", " ", text).strip() # Collapses multiple spaces, tabs, newlines into a single space
    return text
```

# 2. tokenization

```python
def tokenize_text(text):
    tokens = text.split()
    return tokens
```

# 3. stopword

```python
nltk.download('stopwords', quiet=True)

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(tokens):
    return [w for w in tokens if w not in stop_words]
```

```python
tokens = ["you", "are", "a", "stupid", "person"]
remove_stopwords(tokens)
```

# 4. lemmatization

```python
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(w, pos='v') for w in tokens]
```

```python
tokens = ["running", "hated", "better"]
lemmatize_text(tokens)
```

```python
def full_preprocess(text):
    if pd.isna(text):
        return ""
    
    text = clean_text(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize_text(tokens)
    
    return " ".join(tokens)
```

```python
df["cleaned_text"] = df["tweet"].apply(full_preprocess)
```

## GloVe

```python
# ===============================
# GloVe — Load Pretrained Vectors
# ===============================
# Download glove.6B.zip from https://nlp.stanford.edu/projects/glove/
# Unzip and place glove.6B.100d.txt under data/external/glove.6B/ (recommended)
# or next to this notebook.

from pathlib import Path

_glove_candidates = (
    Path("../data/external/glove.6B/glove.6B.100d.txt"),
    Path("../data/external/glove.6B.100d.txt"),
    Path("glove.6B.100d.txt"),
)
glove_path = next((p.resolve() for p in _glove_candidates if p.is_file()), None)
if glove_path is None:
    raise FileNotFoundError(
        "GloVe file not found. Download glove.6B.zip from "
        "https://nlp.stanford.edu/projects/glove/ then extract so you have "
        "HateSpeech-detection/data/external/glove.6B/glove.6B.100d.txt "
        "(or put glove.6B.100d.txt in the notebooks folder)."
    )

glove_embeddings = {}

with open(glove_path, encoding="utf8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        glove_embeddings[word] = vector

print("Loaded", len(glove_embeddings), "GloVe word vectors")
```

```python
# ===============================
# Test Word Vector
# ===============================

word = "trash"

if word in glove_embeddings:
    vector = glove_embeddings[word]
    print("GloVe:", vector[:5])  # show first 5 numbers
else:
    print(word, "not in GloVe vocabulary")
```

```python
# ===============================
# Convert Sentences to GloVe Vectors
# ===============================

def sentence_to_glove_vector(sentence, embeddings, dim=100):
    words = sentence.split()
    
    vectors = []
    
    for word in words:
        if word in embeddings:
            vectors.append(embeddings[word])
    
    if len(vectors) == 0:
        return np.zeros(dim)
    
    return np.mean(vectors, axis=0)


glove_sentence_vectors = np.array([
    sentence_to_glove_vector(text, glove_embeddings)
    for text in df["cleaned_text"]
])

print("GloVe sentence vectors shape:", glove_sentence_vectors.shape)
```

```python
def sentence_embedding(tokens, model, dim=100):
    vectors = [model.wv[w] for w in tokens if w in model.wv]
    if not vectors:
        return np.zeros(dim)
    return np.mean(vectors, axis=0)

embeddings = np.array([sentence_embedding(t, model) for t in X_train["cleaned_text"].astype(str).str.split()])

print(f"Shape de la matrice d'embeddings : {embeddings.shape}")
print(f"→ {embeddings.shape[0]} phrases × {embeddings.shape[1]} dimensions")
print(f"\nExemple vecteur[0] (10 premières valeurs) :")
print(embeddings[0][:10].round(4))
```

```python
X_train_w2v = np.array([
    sentence_embedding(t, w2v_model)
    for t in X_train["cleaned_text"].str.split()
])

X_test_w2v = np.array([
    sentence_embedding(t, w2v_model)
    for t in X_test["cleaned_text"].str.split()
])
```

## TF_IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_features=10000
)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train["cleaned_text"])
X_test_tfidf  = tfidf_vectorizer.transform(X_test["cleaned_text"])

# Display shape
print("TF-IDF matrix shape:", X_train_tfidf.shape)
```

## Combining the two

```python
from scipy.sparse import hstack, csr_matrix

X_train_hybrid = hstack(
    [X_train_tfidf, csr_matrix(X_train_w2v)],
    format="csr"
)

X_test_hybrid = hstack(
    [X_test_tfidf, csr_matrix(X_test_w2v)],
    format="csr"
)
```

## SVM Optimized

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix
# ===============================
# Oversampling
# ===============================
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(
    X_train_hybrid,
    y_train
)
print("Before Oversampling:")
print(y_train.value_counts())

print("\nAfter Oversampling:")
print(pd.Series(y_train_resampled).value_counts())

# ===============================
# Train SVM Model
# ===============================
svm_model = LinearSVC(
    C=0.1, #for stronger regularization and less overfitting
    random_state=42
)
svm_model.fit(X_train_resampled, y_train_resampled)
#svm_model.fit(X_train, y_train)
```

```python
# ===============================
# Predictions
# ===============================
y_pred = svm_model.predict(X_test_hybrid)

# ===============================
# Evaluation
# ===============================
print("\nAccuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report:\n")
print(classification_report(
    y_test,
    y_pred,
    target_names=[
        "hate_speech",
        "offensive_language",
        "neither"
    ]
))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```
