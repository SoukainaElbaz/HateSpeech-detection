# 01 — Exploration des données
EDA : distribution des classes, longueur des textes, mots fréquents.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset
df = pd.read_csv('../data/raw/HateSpeechData.csv')
print(df.shape)
df.head()
```

```python
# Distribution des classes
df['class'].value_counts().plot(kind='bar', color=['steelblue', 'tomato'])

plt.title('Distribution des classes')
plt.xlabel('Label')
plt.ylabel("Nombre d'exemples")

plt.tight_layout()
plt.show()
```

```python
# Longueur des textes
df['text_length'] = df['tweet'].apply(lambda x: len(str(x).split()))
df.groupby('class')['text_length'].hist(bins=40, alpha=0.7)
plt.title('Distribution de la longueur des textes par classe')
plt.xlabel('Nombre de mots')
plt.show()
df['text_length'].describe()
```

```python
# Valeurs manquantes
print('Valeurs manquantes :')
print(df.isnull().sum())

# Doublons
print(f'\nDoublons : {df.duplicated().sum()}')
```
