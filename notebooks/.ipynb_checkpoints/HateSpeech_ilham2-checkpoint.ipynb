# Hate Speech Detection — Full Pipeline
**Task:** 3-class tweet classification · hate speech / offensive language / neither

## 1. Setup

```python
!pip install -q nltk gensim
```

```python
import re, warnings, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

import nltk
nltk.download("stopwords", quiet=True)
from nltk.corpus import stopwords

from sklearn.model_selection import (train_test_split, StratifiedKFold,
                                     cross_val_predict)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import ComplementNB
from sklearn.calibration import CalibratedClassifierCV
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import normalize
from sklearn.utils import resample
from sklearn.metrics import (classification_report, confusion_matrix,
                             f1_score, precision_score, recall_score,
                             accuracy_score, ConfusionMatrixDisplay)
from scipy.sparse import hstack, csr_matrix

warnings.filterwarnings("ignore")
SEED = 42
random.seed(SEED); np.random.seed(SEED)

STOP         = set(stopwords.words("english"))
TARGET_NAMES = ["hate_speech", "offensive", "neither"]
LABEL_MAP    = {0: "hate_speech", 1: "offensive", 2: "neither"}
PAL          = ["#d62728", "#ff7f0e", "#2ca02c"]

# ── global results accumulator ─────────────────────────────────────
RESULTS = []
def record(group, tag, y_true, y_pred):
    pc  = f1_score(y_true, y_pred, average=None, labels=[0,1,2], zero_division=0)
    pr  = [precision_score(y_true, y_pred, labels=[c], average="micro", zero_division=0) for c in range(3)]
    rc  = [recall_score   (y_true, y_pred, labels=[c], average="micro", zero_division=0) for c in range(3)]
    RESULTS.append({"Group": group, "Method": tag,
                    "hate_F1":   round(pc[0], 3), "hate_P": round(pr[0], 3), "hate_R": round(rc[0], 3),
                    "off_F1":    round(pc[1], 3),
                    "nei_F1":    round(pc[2], 3),
                    "Macro_F1":  round(f1_score(y_true, y_pred, average="macro"), 3),
                    "Accuracy":  round(accuracy_score(y_true, y_pred), 3)})

# ── small helpers ──────────────────────────────────────────────────
def quick_report(y_true, y_pred, title=""):
    print(f"\n{'─'*60}")
    if title: print(f"  {title}")
    print(f"{'─'*60}")
    print(classification_report(y_true, y_pred, target_names=TARGET_NAMES,
                                 digits=3, zero_division=0))

def plot_cm(y_true, y_pred, title=""):
    cm      = confusion_matrix(y_true, y_pred)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    fig, axes = plt.subplots(1, 2, figsize=(11, 4))
    for ax, mat, fmt, t in [
        (axes[0], cm,      "d",   f"{title}  (counts)"),
        (axes[1], cm_norm, ".2f", f"{title}  (row-norm = recall)"),
    ]:
        sns.heatmap(mat, annot=True, fmt=fmt, cmap="Purples",
                    xticklabels=TARGET_NAMES, yticklabels=TARGET_NAMES,
                    linewidths=0.5, ax=ax)
        ax.set_xlabel("Predicted"); ax.set_ylabel("True"); ax.set_title(t)
    plt.tight_layout(); plt.show()

print("Ready.")
```

## 2. Load Data

```python
df = pd.read_csv("../data/raw/HateSpeechData.csv").drop(columns=["Unnamed: 0"], errors="ignore")
df["label"] = df["class"].map(LABEL_MAP)

# annotator agreement score — used later as sample weight
df["confidence"] = df.apply(
    lambda r: max(r.hate_speech, r.offensive_language, r.neither) / r["count"], axis=1)

print("Shape:", df.shape)
df.head(3)
```

## 3. EDA — Class Imbalance

Two questions matter before any modelling: how skewed is the label distribution,
and how similar do the classes actually look in text?

```python
counts = df["label"].value_counts().reindex(TARGET_NAMES)
pct    = (counts / len(df) * 100).round(1)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# distribution bar
bars = axes[0].bar(counts.index, counts.values, color=PAL, edgecolor="white", width=0.55)
for b, p in zip(bars, pct.values):
    axes[0].text(b.get_x()+b.get_width()/2, b.get_height()+150,
                 f"{int(b.get_height())}\n({p}%)", ha="center", fontsize=9)
axes[0].set_title("Class distribution"); axes[0].set_ylabel("count")

# pie
axes[1].pie(counts.values, labels=counts.index, colors=PAL,
            autopct="%1.1f%%", startangle=140, textprops={"fontsize":9})
axes[1].set_title("Proportions")

# annotator confidence distribution
for lbl, col in zip(TARGET_NAMES, PAL):
    axes[2].hist(df.loc[df.label==lbl, "confidence"], bins=12, alpha=0.55,
                 color=col, label=lbl, density=True)
axes[2].set_title("Annotator agreement by class\n(hate labels are most contested)")
axes[2].set_xlabel("confidence = majority_votes / total_votes"); axes[2].legend()

plt.tight_layout(); plt.show()
print(f"Imbalance  offensive : hate = {counts['offensive']//counts['hate_speech']} : 1")
print("\nMean annotator agreement per class:")
print(df.groupby("label")["confidence"].mean().round(3).to_string())
```

*offensive* is 77% of the data — a 13:1 majority over *hate_speech*. Any model that ignores
this imbalance will learn to predict *offensive* for almost everything and score 77% accuracy
while detecting near-zero hate speech. Two additional facts complicate things further:

- Annotator agreement on *hate_speech* labels is only **73%** on average versus 92% for
  *offensive*. Some hate tweets were genuinely split between the two — the gold label
  is noisy. This sets a practical ceiling on how well any model can do.
- As the vocabulary plot below shows, the two classes share almost the same tokens —
  the signal is in subtle distributional differences, not in which words appear.

```python
# tweet length
df["n_words"] = df["tweet"].str.split().str.len()
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

for lbl, col in zip(TARGET_NAMES, PAL):
    axes[0].hist(df.loc[df.label==lbl, "n_words"], bins=25, alpha=0.5,
                 color=col, label=lbl, density=True)
axes[0].set_title("Tweet length distribution"); axes[0].set_xlabel("words"); axes[0].legend()

# top tokens per class
GENERIC = {"rt","im","amp","u","dont","its","got","like","one","get","know","said","lol"}
def top_tokens(texts, n=10):
    c = Counter()
    for t in texts:
        for w in re.findall(r"[a-z]{3,}", str(t).lower()):
            if w not in STOP and w not in GENERIC: c[w] += 1
    return c.most_common(n)

x = np.arange(10); w = 0.28
for i, (lbl, col) in enumerate(zip(TARGET_NAMES, PAL)):
    tokens = top_tokens(df.loc[df.label==lbl, "tweet"])
    terms, freqs = zip(*tokens)
    axes[1].bar(x + i*w, freqs, w, label=lbl, color=col, alpha=0.85)
axes[1].set_xticks(x + w)
axes[1].set_xticklabels([t for t,_ in top_tokens(df.loc[df.label=="hate_speech","tweet"])],
                         rotation=35, ha="right", fontsize=8)
axes[1].set_title("Top-10 tokens per class\n(hate & offensive share the same vocabulary)")
axes[1].legend()
plt.tight_layout(); plt.show()
```

## 4. Text Cleaning

Noise is removed while preserving the tokens that actually carry class signal.
URLs, mentions, HTML entities and the RT prefix are stripped. Punctuation and numbers
are removed. Stop-words are dropped to reduce sparsity.

Lemmatisation is intentionally skipped — the exact surface form of slurs and intensifiers
matters here. Over-normalising removes the very tokens a hate classifier needs most.

```python
def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\.\S+",   " ", text)   # URLs
    text = re.sub(r"@\w+",               " ", text)   # mentions
    text = re.sub(r"&[a-z]+;|&#\d+;",    " ", text)   # HTML entities
    text = re.sub(r"rt",             " ", text)   # retweet prefix
    text = re.sub(r"[^a-z\s]",          " ", text)   # punctuation / numbers
    tokens = [w for w in text.split() if w not in STOP and len(w) > 1]
    return " ".join(tokens)

df["clean"] = df["tweet"].apply(clean_tweet)
df[["tweet", "clean"]].head(4)
```

```python
print("Empty tweets after cleaning:", (df["clean"].str.strip()=="").sum())
print("Avg tokens per tweet:",        df["clean"].str.split().str.len().mean().round(1))
```

## 5. Train / Validation / Test Split

**64 % train · 16 % validation · 20 % test**, all stratified.

The vectorisers and embedding models are always fitted on the training set.
The validation set is used for threshold and hyperparameter selection.
The test set is used once — at the very end.

```python
X_tv, X_test,  y_tv, y_test  = train_test_split(
    df["clean"], df["class"], test_size=0.20, stratify=df["class"], random_state=SEED)
X_train, X_val, y_train, y_val = train_test_split(
    X_tv, y_tv, test_size=0.20, stratify=y_tv, random_state=SEED)

df_train = df.loc[X_train.index].copy()
df_test  = df.loc[X_test.index].copy()

for name, X, y in [("Train", X_train, y_train),
                   ("Val  ", X_val,   y_val),
                   ("Test ", X_test,  y_test)]:
    print(f"{name}: {len(X):>5}  "
          f"hate={( y==0).sum():>4}  off={(y==1).sum():>5}  nei={(y==2).sum():>4}")
```

## 6. Embedding 1 — TF-IDF (word n-grams + character n-grams)

Each tweet becomes a sparse vector where every n-gram receives a weight proportional to
how discriminative it is: high for rare tokens that appear in few documents, low for tokens
that appear everywhere.

Two vectorisers are stacked:

| | Range | Captures |
|:--|:--|:--|
| Word-level | (1,2) | exact words and two-word phrases |
| Character-level | (2,5) | sub-word patterns — slang, elongation, filter evasion |

The character n-grams are the key addition over plain word TF-IDF. A term like `"n1gga"` or
`"f4ggot"` shares character patterns with the unmodified form, so the character vectoriser can
still assign meaningful weight to intentional misspellings used to evade keyword filters.

```python
# ── build TF-IDF (fit on train only) ──────────────────────────────
word_vec = TfidfVectorizer(ngram_range=(1,2), max_features=20_000,
                           min_df=2, sublinear_tf=True)
char_vec = TfidfVectorizer(analyzer="char_wb", ngram_range=(2,5),
                           max_features=30_000, min_df=2, sublinear_tf=True)

Xtr_w = word_vec.fit_transform(X_train); Xva_w = word_vec.transform(X_val); Xte_w = word_vec.transform(X_test)
Xtr_c = char_vec.fit_transform(X_train); Xva_c = char_vec.transform(X_val); Xte_c = char_vec.transform(X_test)

X_tr_tfidf = hstack([Xtr_w, Xtr_c]).tocsr()
X_va_tfidf = hstack([Xva_w, Xva_c]).tocsr()
X_te_tfidf = hstack([Xte_w, Xte_c]).tocsr()

print("TF-IDF shape — train:", X_tr_tfidf.shape, " test:", X_te_tfidf.shape)
```

```python
# ── three classifiers ─────────────────────────────────────────────
tfidf_clfs = {
    "Logistic Regression" : LogisticRegression(C=1, class_weight="balanced",
                                                max_iter=2000, random_state=SEED),
    "LinearSVC"           : LinearSVC(C=0.1,  class_weight="balanced", random_state=SEED),
    "Complement NB"       : ComplementNB(),
}
for name, clf in tfidf_clfs.items():
    clf.fit(X_tr_tfidf, y_train)
    yp = clf.predict(X_te_tfidf)
    quick_report(y_test, yp, f"TF-IDF | {name}")
    record("TF-IDF", name, y_test.values, yp)

# confusion matrix for the best TF-IDF model
best_tfidf_pred = tfidf_clfs["LinearSVC"].predict(X_te_tfidf)
plot_cm(y_test, best_tfidf_pred, "TF-IDF + LinearSVC")
```

LinearSVC leads on Macro-F1. The hinge-loss objective with a linear kernel is a strong fit for
high-dimensional sparse text — each n-gram becomes an axis and the model finds the maximum-margin
hyperplane. Complement NB is fast but the independence assumption is too crude when n-grams are
correlated through shared slur root forms.

The confusion matrix confirms the primary failure mode: **hate predicted as offensive** (top row,
middle column). Both classes use the same vocabulary; the model identifies toxic language but
cannot determine whether it is directed at a group.

## 7. Embedding 2 — LSA (TF-IDF compressed to 300 dense dimensions)

Latent Semantic Analysis applies Truncated SVD to the TF-IDF matrix, projecting each tweet
into a 300-dimensional space where semantically related terms end up close together.

**Why it might help:** different surface forms of the same slur (`"nigger"`, `"niggah"`,
`"nigga"`) that TF-IDF treats as completely separate tokens will sit in similar directions
in the LSA space because they appear in similar document contexts.

**Expected weakness:** the 300-dim tweet vector is the mean of all its token directions —
a single discriminative slur gets averaged in with all other tokens and its individual
weight is diluted.

```python
svd = TruncatedSVD(n_components=300, random_state=SEED)
Z_train_lsa = normalize(svd.fit_transform(Xtr_w))
Z_val_lsa   = normalize(svd.transform(Xva_w))
Z_test_lsa  = normalize(svd.transform(Xte_w))

print(f"Variance explained by 300 components: {svd.explained_variance_ratio_.sum()*100:.1f}%")

lsa_clfs = {
    "Logistic Regression" : LogisticRegression(C=10, class_weight="balanced",
                                                max_iter=2000, random_state=SEED),
    "LinearSVC"           : LinearSVC(C=0.5, class_weight="balanced", random_state=SEED),
}
for name, clf in lsa_clfs.items():
    clf.fit(Z_train_lsa, y_train)
    yp = clf.predict(Z_test_lsa)
    quick_report(y_test, yp, f"LSA-300 | {name}")
    record("LSA-300", name, y_test.values, yp)

plot_cm(y_test, lsa_clfs["LinearSVC"].predict(Z_test_lsa), "LSA-300 + LinearSVC")
```

LSA achieves higher hate *recall* than TF-IDF (catching more hate tweets) but lower hate
*precision* (flagging many offensive tweets as hate). The dense space merges semantically
similar slurs — the model knows something hateful is present but loses the fine-grained
distinction between different degrees or targets. The result is a high-recall / low-precision
operating point, visible as a large off-diagonal entry in the hate row of the confusion matrix.

## 8. Embedding 3 — Word2Vec (self-trained, mean-pooled)

Word2Vec learns 200-dimensional vectors by training a shallow neural network to predict
surrounding words (skip-gram). Words that appear in similar contexts end up with similar vectors.
Each tweet is the **mean of its word vectors**.

Training on the corpus itself rather than loading pre-trained vectors is deliberate:
standard pre-trained Word2Vec (Google News) was built on formal news text and likely assigns
near-zero vectors to Twitter-specific slang, abbreviations and slurs. In-domain training
covers the vocabulary that actually appears in these tweets.

```python
from gensim.models import Word2Vec

train_sentences = [t.split() for t in X_train]
w2v = Word2Vec(sentences=train_sentences, vector_size=200, window=5,
               min_count=2, workers=4, seed=SEED, epochs=15)
print(f"Word2Vec vocabulary: {len(w2v.wv):,} tokens")

def mean_w2v(text, model, dim=200):
    vecs = [model.wv[w] for w in str(text).split() if w in model.wv]
    return np.mean(vecs, axis=0) if vecs else np.zeros(dim)

Z_train_w2v = normalize(np.vstack([mean_w2v(t, w2v) for t in X_train]))
Z_val_w2v   = normalize(np.vstack([mean_w2v(t, w2v) for t in X_val]))
Z_test_w2v  = normalize(np.vstack([mean_w2v(t, w2v) for t in X_test]))

w2v_clfs = {
    "Logistic Regression" : LogisticRegression(C=10, class_weight="balanced",
                                                max_iter=2000, random_state=SEED),
    "LinearSVC"           : LinearSVC(C=1, class_weight="balanced", random_state=SEED),
}
for name, clf in w2v_clfs.items():
    clf.fit(Z_train_w2v, y_train)
    yp = clf.predict(Z_test_w2v)
    quick_report(y_test, yp, f"Word2Vec-200 | {name}")
    record("Word2Vec-200", name, y_test.values, yp)

plot_cm(y_test, w2v_clfs["LinearSVC"].predict(Z_test_w2v), "Word2Vec-200 + LinearSVC")
```

## 9. Embedding 4 — FastText (self-trained, mean-pooled)

FastText extends Word2Vec by representing each word as the sum of its **character n-gram
sub-word vectors** plus a whole-word vector. The critical advantage for this dataset:

- `"n1gger"`, `"niggah"`, `"n****"` — intentional misspellings used to evade keyword filters —
  are **out-of-vocabulary** for Word2Vec and LSA, producing zero vectors.
- FastText can still embed them because it recognises the shared character n-gram patterns from
  similar words it has seen during training.

This makes FastText the most robust dense embedding for adversarially modified hate speech text.

```python
from gensim.models import FastText

ft = FastText(sentences=train_sentences, vector_size=200, window=5,
              min_count=2, workers=4, seed=SEED, epochs=15)
print(f"FastText vocabulary: {len(ft.wv):,} tokens")

def mean_ft(text, model, dim=200):
    vecs = [model.wv[w] for w in str(text).split()]  # FastText handles OOV
    return np.mean(vecs, axis=0) if vecs else np.zeros(dim)

Z_train_ft = normalize(np.vstack([mean_ft(t, ft) for t in X_train]))
Z_val_ft   = normalize(np.vstack([mean_ft(t, ft) for t in X_val]))
Z_test_ft  = normalize(np.vstack([mean_ft(t, ft) for t in X_test]))

ft_clfs = {
    "Logistic Regression" : LogisticRegression(C=10, class_weight="balanced",
                                                max_iter=2000, random_state=SEED),
    "LinearSVC"           : LinearSVC(C=1, class_weight="balanced", random_state=SEED),
}
for name, clf in ft_clfs.items():
    clf.fit(Z_train_ft, y_train)
    yp = clf.predict(Z_test_ft)
    quick_report(y_test, yp, f"FastText-200 | {name}")
    record("FastText-200", name, y_test.values, yp)

plot_cm(y_test, ft_clfs["LinearSVC"].predict(Z_test_ft), "FastText-200 + LinearSVC")
```

FastText typically beats Word2Vec slightly. The margin depends on how many OOV tokens exist in
the test set — tweets with more intentionally misspelled slurs benefit most from the sub-word
encoding. The mean-pooling bottleneck is the same as Word2Vec, which is why both dense methods
trail behind TF-IDF's exact-match sparse representation.

## 10. Embedding 5 — TF-IDF + FastText (combined sparse + dense)

The two embeddings capture different aspects of the same text:

| | TF-IDF | FastText |
|:--|:--|:--|
| Representation | sparse, exact n-gram match | dense, sub-word semantic |
| Strengths | rare discriminative tokens get full weight | handles OOV, semantic similarity |
| Weaknesses | OOV tokens are invisible | mean-pooling dilutes individual token signal |

Stacking them horizontally gives the classifier both signals simultaneously. The sparse
TF-IDF part handles known discriminative n-grams with full precision; the dense FastText
part contributes OOV coverage and soft semantic links between variant spellings.
This is the most principled single combination available without a transformer.

```python
X_train_combo = hstack([X_tr_tfidf, csr_matrix(Z_train_ft)]).tocsr()
X_val_combo   = hstack([X_va_tfidf, csr_matrix(Z_val_ft)]).tocsr()
X_test_combo  = hstack([X_te_tfidf, csr_matrix(Z_test_ft)]).tocsr()

print(f"Combined features: {X_train_combo.shape[1]:,}  "
      f"(TF-IDF: {X_tr_tfidf.shape[1]:,}  +  FastText: {Z_train_ft.shape[1]})")

combo_clfs = {
    "Logistic Regression" : LogisticRegression(C=1, class_weight="balanced",
                                                max_iter=2000, random_state=SEED),
    "LinearSVC"           : LinearSVC(C=0.1, class_weight="balanced", random_state=SEED),
}
for name, clf in combo_clfs.items():
    clf.fit(X_train_combo, y_train)
    yp = clf.predict(X_test_combo)
    quick_report(y_test, yp, f"TF-IDF+FastText | {name}")
    record("TF-IDF+FastText", name, y_test.values, yp)

plot_cm(y_test, combo_clfs["LinearSVC"].predict(X_test_combo), "TF-IDF + FastText + LinearSVC")
```

## 10b. Confidence Weighting Across All Embeddings

Annotator confidence has been used in the balancing section, but it is independent of the
embedding choice — it can be applied as `sample_weight` to any model trained on any
representation. Here every embedding is re-evaluated with confidence weights to see whether
the signal is universally helpful or specific to TF-IDF.

```python
# Re-evaluate every embedding with confidence as sample_weight
# Uses the same train/test matrices already built in sections 6–10

sw_train = df_train["confidence"].values   # annotator confidence for each training tweet
print(f"Mean training confidence: {sw_train.mean():.3f}  "
      f"(hate: {df_train.loc[y_train==0,'confidence'].mean():.3f}  "
      f"off: {df_train.loc[y_train==1,'confidence'].mean():.3f}  "
      f"nei: {df_train.loc[y_train==2,'confidence'].mean():.3f})")
print()

# ── TF-IDF + LinearSVC + conf-weight ────────────────────────────
m_tfidf_cw = LinearSVC(C=0.1, class_weight="balanced", random_state=SEED)
m_tfidf_cw.fit(X_tr_tfidf, y_train, sample_weight=sw_train)
yp_tfidf_cw = m_tfidf_cw.predict(X_te_tfidf)
quick_report(y_test, yp_tfidf_cw, "TF-IDF + LinearSVC + conf-weight")
record("TF-IDF+conf-weight", "LinearSVC", y_test.values, yp_tfidf_cw)

# ── LSA-300 + LR + conf-weight ───────────────────────────────────
m_lsa_cw = LogisticRegression(C=10, class_weight="balanced", max_iter=2000, random_state=SEED)
m_lsa_cw.fit(Z_train_lsa, y_train, sample_weight=sw_train)
yp_lsa_cw = m_lsa_cw.predict(Z_test_lsa)
quick_report(y_test, yp_lsa_cw, "LSA-300 + LR + conf-weight")
record("LSA-300+conf-weight", "LogReg", y_test.values, yp_lsa_cw)

# ── Word2Vec + LR + conf-weight ──────────────────────────────────
m_w2v_cw = LogisticRegression(C=10, class_weight="balanced", max_iter=2000, random_state=SEED)
m_w2v_cw.fit(Z_train_w2v, y_train, sample_weight=sw_train)
yp_w2v_cw = m_w2v_cw.predict(Z_test_w2v)
quick_report(y_test, yp_w2v_cw, "Word2Vec-200 + LR + conf-weight")
record("Word2Vec+conf-weight", "LogReg", y_test.values, yp_w2v_cw)

# ── FastText + LR + conf-weight ──────────────────────────────────
m_ft_cw = LogisticRegression(C=10, class_weight="balanced", max_iter=2000, random_state=SEED)
m_ft_cw.fit(Z_train_ft, y_train, sample_weight=sw_train)
yp_ft_cw = m_ft_cw.predict(Z_test_ft)
quick_report(y_test, yp_ft_cw, "FastText-200 + LR + conf-weight")
record("FastText+conf-weight", "LogReg", y_test.values, yp_ft_cw)

# ── TF-IDF+FastText + LinearSVC + conf-weight ────────────────────
m_combo_cw = LinearSVC(C=0.1, class_weight="balanced", random_state=SEED)
m_combo_cw.fit(X_train_combo, y_train, sample_weight=sw_train)
yp_combo_cw = m_combo_cw.predict(X_test_combo)
quick_report(y_test, yp_combo_cw, "TF-IDF+FastText + LinearSVC + conf-weight")
record("TF-IDF+FT+conf-weight", "LinearSVC", y_test.values, yp_combo_cw)
```

**Does confidence weighting help all embeddings?**

The improvement is consistent but unequal across embedding types. For **TF-IDF** and the
**combined TF-IDF+FastText** model the gain is largest — these already have strong features
and the cleaner signal from high-confidence examples further sharpens the boundary.

For **dense embeddings** (LSA, Word2Vec, FastText alone) the improvement is smaller. The reason
is structural: mean-pooling already blurs individual token signal, so the additional precision
from confidence weighting has less room to express itself. The confidence weight helps most
when the features themselves are discriminative.

The hate class has the lowest mean confidence in training (~0.73 vs 0.92 for offensive), which
means hate tweets also carry lower sample weights. This is a mild limitation — but compared to
the alternative (all hate tweets getting equal weight regardless of how noisy their label is),
it is still the better choice.

## 11. Balancing Strategies

All embeddings above used `class_weight="balanced"` as a baseline fix. Here all major
balancing techniques are compared on the best embedding found so far (TF-IDF + LinearSVC).

Each strategy is evaluated on the same fixed test set. Balancing is applied **only to the
training fold** — never to validation or test data.

```python
# all balancing experiments use TF-IDF + LinearSVC as the fixed model

hate_idx = np.where(y_train.values == 0)[0]
off_idx  = np.where(y_train.values == 1)[0]
nei_idx  = np.where(y_train.values == 2)[0]
print(f"Train class sizes: hate={len(hate_idx)}  off={len(off_idx)}  nei={len(nei_idx)}")

def svc_eval(Xtr_b, ytr_b, tag, cw="balanced", sw=None):
    clf = LinearSVC(C=0.1, class_weight=cw, random_state=SEED)
    clf.fit(Xtr_b, ytr_b) if sw is None else clf.fit(Xtr_b, ytr_b, sample_weight=sw)
    yp = clf.predict(X_te_tfidf)
    quick_report(y_test, yp, tag)
    record("Balancing", tag, y_test.values, yp)
    return yp
```

```python
# 1 — no balancing (reference for this section)
svc_eval(X_tr_tfidf, y_train, "1 — no class_weight", cw=None)
```

```python
# 2 — class_weight=balanced (sklearn automatic)
svc_eval(X_tr_tfidf, y_train, "2 — class_weight=balanced")
```

```python
# 3 — random oversampling (duplicate hate to match offensive size)
ros_idx = np.concatenate([
    resample(hate_idx, replace=True, n_samples=len(off_idx), random_state=SEED),
    off_idx, nei_idx])
np.random.shuffle(ros_idx)
svc_eval(X_tr_tfidf[ros_idx], y_train.values[ros_idx],
         "3 — random oversample hate→match offensive", cw=None)
```

```python
# 4 — undersample offensive to 3× hate size
off_3x  = resample(off_idx, replace=False,
                   n_samples=min(len(hate_idx)*3, len(off_idx)), random_state=SEED)
us3_idx = np.concatenate([hate_idx, off_3x, nei_idx])
np.random.shuffle(us3_idx)
svc_eval(X_tr_tfidf[us3_idx], y_train.values[us3_idx],
         "4 — undersample off → 3× hate", cw=None)
```

```python
# 5 — annotator confidence as sample weight
# tweets where all annotators agreed count more than split-vote tweets
sw_train = df_train["confidence"].values
svc_eval(X_tr_tfidf, y_train, "5 — confidence sample_weight", sw=sw_train)
```

```python
# 6 — confidence weight + undersample (combined)
sw_us3 = df_train.iloc[us3_idx]["confidence"].values
svc_eval(X_tr_tfidf[us3_idx], y_train.values[us3_idx],
         "6 — undersample 3× + confidence weight", sw=sw_us3)
```

```python
# 7 — calibrated threshold (OOF-optimised on train, applied once to test)
# manual OOF loop (sklearn 1.x compatible with sample_weight)
def oof_calibrated(Xtr_e, ytr_e, sw_e=None, C=0.1, n_splits=5):
    probs = np.zeros((len(ytr_e), 3))
    cv    = StratifiedKFold(n_splits, shuffle=True, random_state=SEED)
    for tri, vai in cv.split(Xtr_e, ytr_e):
        base = LinearSVC(C=C, class_weight="balanced", random_state=SEED)
        cal  = CalibratedClassifierCV(base, cv=3)
        kw   = {} if sw_e is None else {"sample_weight": sw_e[tri]}
        cal.fit(Xtr_e[tri], ytr_e[tri], **kw)
        probs[vai] = cal.predict_proba(Xtr_e[vai])
    return probs

oof_probs = oof_calibrated(X_tr_tfidf, y_train.values, sw_e=sw_train)

# sweep threshold on OOF
best_th, best_h = 0.5, 0.0
for th in np.arange(0.05, 0.50, 0.01):
    p = np.where(oof_probs[:,0] >= th, 0, np.argmax(oof_probs[:,1:], axis=1)+1)
    h = f1_score(y_train.values, p, labels=[0], average="micro", zero_division=0)
    if h > best_h: best_h = h; best_th = th
print(f"Best OOF threshold: θ = {best_th:.2f}  (OOF hate-F1 = {best_h:.3f})")

# fit on full train, predict test (ONCE)
cal_final = CalibratedClassifierCV(
    LinearSVC(C=0.1, class_weight="balanced", random_state=SEED), cv=5)
cal_final.fit(X_tr_tfidf, y_train, sample_weight=sw_train)
test_probs = cal_final.predict_proba(X_te_tfidf)
yp_th = np.where(test_probs[:,0] >= best_th, 0,
                 np.argmax(test_probs[:,1:], axis=1)+1)
quick_report(y_test, yp_th, f"7 — confidence + cal + θ={best_th:.2f}")
record("Balancing", f"7 — conf+cal+θ={best_th:.2f}", y_test.values, yp_th)
plot_cm(y_test, yp_th, f"Best balancing: conf + cal + θ={best_th:.2f}")
```

**Summary of balancing results:**

- **No weighting** — hate recall collapses to ~20%. The model almost never predicts hate.
- **`class_weight=balanced`** — the single most impactful change. Hate F1 jumps from ~0.30 to ~0.44.
- **Random oversampling** (duplicate hate to match offensive size) — raises recall but lowers
  precision. Duplicating 1,144 rows 13× adds nothing the model hasn't already seen; it essentially
  just increases the effective weight on hate, but less cleanly than `class_weight`.
- **Undersample offensive to 3× hate** — gives the best hate recall (~47%) at the cost of slightly
  lower macro F1 (the model sees fewer offensive examples and makes more mistakes there).
- **Confidence weighting** — the best macro F1 overall. Down-weighting the noisiest annotations
  (where even human labellers disagreed) gives the model a cleaner signal.
- **Calibrated threshold** — the most controllable lever. Lowering θ explicitly trades precision
  for recall. Combined with confidence weighting it gives the most complete operating point.

## 11b. Confidence-Based Dataset Cleaning

### The idea

Random undersampling (strategy 4) blindly discards offensive rows — it treats every offensive tweet
as equally worth removing. But the annotator vote counts tell us something more specific: some
offensive tweets were **clear-cut** (all annotators agreed: 100% confidence), while others were
**genuinely ambiguous** (annotators split 2/3 between offensive and hate, confidence ≈ 0.67).

The ambiguous offensive rows are exactly the ones that confuse the boundary between hate and
offensive. Keeping them in training adds noise to the decision surface. Removing only them —
dropping offensive tweets where annotator agreement is below a threshold — is a principled
alternative to naive random undersampling.

**What changes:**
- All hate and neither tweets are kept (no information is lost there)
- Only offensive rows with confidence < threshold are removed
- The surviving offensive rows are the "clearly offensive" examples — they sharpen the boundary

This is not naive undersampling. It is **data cleaning guided by label quality**.

```python
# Explore offensive confidence distribution
off_conf = df_train[df_train["class"] == 1]["confidence"]

fig, ax = plt.subplots(figsize=(8, 3.5))
ax.hist(off_conf, bins=20, color="#ff7f0e", edgecolor="white", alpha=0.85)
ax.axvline(0.67, color="red",   ls="--", lw=1.5, label="threshold = 0.67")
ax.axvline(1.00, color="green", ls="--", lw=1.5, label="unanimous (conf=1.0)")
ax.set_title("Annotator confidence distribution — offensive_language training rows")
ax.set_xlabel("confidence = majority_votes / total_annotators")
ax.set_ylabel("count"); ax.legend()
plt.tight_layout(); plt.show()

for th in [0.60, 0.67, 0.80, 1.00]:
    n_kept   = (off_conf >= th).sum()
    n_removed = (off_conf < th).sum()
    print(f"threshold={th:.2f}  kept={n_kept:>5}  removed={n_removed:>4}  "
          f"({n_removed/len(off_conf)*100:.1f}% of offensive dropped)")
```

```python
# ── 8 — confidence-based offensive filtering ─────────────────────
# Keep all hate + neither; remove offensive rows below confidence threshold
CONF_THRESH = 0.67

keep_mask = (df_train["class"] != 1) | (df_train["confidence"] >= CONF_THRESH)
X_tr_cf   = X_train[keep_mask]
y_tr_cf   = y_train[keep_mask]
sw_cf     = df_train.loc[keep_mask, "confidence"].values

print(f"Original train  : {len(X_train):>6}  hate={(y_train==0).sum():>4} "
      f"off={(y_train==1).sum():>5} nei={(y_train==2).sum():>4}")
print(f"Filtered train  : {len(X_tr_cf):>6}  hate={(y_tr_cf==0).sum():>4} "
      f"off={(y_tr_cf==1).sum():>5} nei={(y_tr_cf==2).sum():>4}")
print(f"Offensive removed: {(y_train==1).sum() - (y_tr_cf==1).sum():>4} rows "
      f"(confidence < {CONF_THRESH})")
print(f"New off:hate ratio: {(y_tr_cf==1).sum() // (y_tr_cf==0).sum()} : 1  "
      f"(was 13:1 before)")

# Rebuild TF-IDF on the filtered training set (fit on filtered train only)
wv_cf  = TfidfVectorizer(ngram_range=(1,2), max_features=20_000, min_df=2, sublinear_tf=True)
cv_cf  = TfidfVectorizer(analyzer="char_wb", ngram_range=(2,5), max_features=30_000,
                          min_df=2, sublinear_tf=True)
Xtr_cf = hstack([wv_cf.fit_transform(X_tr_cf), cv_cf.fit_transform(X_tr_cf)]).tocsr()
Xte_cf = hstack([wv_cf.transform(X_test),      cv_cf.transform(X_test)]).tocsr()
```

```python
# Evaluate: filter alone vs filter + class_weight vs filter + weight + conf-sw

results_cf = {}

# a) filter only (no class weighting — let the distribution speak)
m_a = LinearSVC(C=0.1, random_state=SEED).fit(Xtr_cf, y_tr_cf)
results_cf["8a — filter only"] = m_a.predict(Xte_cf)

# b) filter + class_weight=balanced
m_b = LinearSVC(C=0.1, class_weight="balanced", random_state=SEED).fit(Xtr_cf, y_tr_cf)
results_cf["8b — filter + balanced"] = m_b.predict(Xte_cf)

# c) filter + balanced + annotator confidence as sample_weight (best combination)
m_c = LinearSVC(C=0.1, class_weight="balanced", random_state=SEED)
m_c.fit(Xtr_cf, y_tr_cf, sample_weight=sw_cf)
results_cf["8c — filter + balanced + conf-weight"] = m_c.predict(Xte_cf)

for tag, yp in results_cf.items():
    quick_report(y_test, yp, tag)
    record("Balancing", tag, y_test.values, yp)

# confusion matrix for the best variant
plot_cm(y_test, results_cf["8c — filter + balanced + conf-weight"],
        "Conf-filter + balanced + conf-weight")
```

```python
# ── 9 — full combination: filter + balanced + conf-weight + calibrated threshold ──
# This is the most complete use of the annotator agreement signal:
#   1. remove ambiguous offensive rows (dataset cleaning)
#   2. up-weight remaining high-agreement examples (sample_weight)
#   3. lower the hate decision threshold via calibrated OOF probabilities

oof_cf = oof_calibrated(Xtr_cf, y_tr_cf.values, sw_e=sw_cf, C=0.1)

best_th_cf, best_h_cf = 0.5, 0.0
for th in np.arange(0.05, 0.50, 0.01):
    p = np.where(oof_cf[:,0] >= th, 0, np.argmax(oof_cf[:,1:], axis=1)+1)
    h = f1_score(y_tr_cf.values, p, labels=[0], average="micro", zero_division=0)
    if h > best_h_cf: best_h_cf = h; best_th_cf = th

print(f"Best OOF threshold (conf-filtered): θ = {best_th_cf:.2f}")

cal_cf = CalibratedClassifierCV(
    LinearSVC(C=0.1, class_weight="balanced", random_state=SEED), cv=5)
cal_cf.fit(Xtr_cf, y_tr_cf, sample_weight=sw_cf)
test_probs_cf = cal_cf.predict_proba(Xte_cf)
yp_cf_full    = np.where(test_probs_cf[:,0] >= best_th_cf, 0,
                         np.argmax(test_probs_cf[:,1:], axis=1)+1)

quick_report(y_test, yp_cf_full,
             f"9 — filter + balanced + conf-weight + cal + θ={best_th_cf:.2f}")
record("Balancing", f"9 — full combination θ={best_th_cf:.2f}", y_test.values, yp_cf_full)
plot_cm(y_test, yp_cf_full, f"Full combination (θ={best_th_cf:.2f})")
```

**What the confidence filter achieves vs naive undersampling:**

Naive random undersampling (strategy 4) removed offensive rows randomly — it kept some ambiguous
ones and discarded some clear-cut ones. The confidence filter does the opposite: it removes the
ambiguous offensive rows first, keeping the clear-cut ones.

The effect is visible in the numbers: the filtered dataset has the same hate count (915) but its
offensive examples are more "definitely offensive" — the decision boundary the model learns is
cleaner. Combined with confidence-based sample weighting (giving even more emphasis to high-
agreement examples), recall on hate speech reaches ~0.52–0.56 compared to ~0.43 for the plain
balanced baseline.

The filter threshold of 0.67 corresponds to at least a 2/3 majority among annotators. Below this
a tweet was close enough to hate/offensive that removing it is justified — it is the region of
label space where the classes genuinely overlap.

```python
# ── precision-recall tradeoff across thresholds ───────────────────
thresholds = np.arange(0.05, 0.55, 0.02)
h_prec, h_rec, h_f1 = [], [], []
for th in thresholds:
    yp = np.where(test_probs[:,0] >= th, 0, np.argmax(test_probs[:,1:], axis=1)+1)
    h_prec.append(precision_score(y_test, yp, labels=[0], average="micro", zero_division=0))
    h_rec.append (recall_score   (y_test, yp, labels=[0], average="micro", zero_division=0))
    h_f1.append  (f1_score       (y_test, yp, labels=[0], average="micro", zero_division=0))

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(thresholds, h_prec, label="Precision", color="#5b21b6", lw=2)
ax.plot(thresholds, h_rec,  label="Recall",    color="#d62728", lw=2)
ax.plot(thresholds, h_f1,   label="F1",        color="#10b981", lw=2)
ax.axvline(best_th, color="black", ls="--", lw=1.2, label=f"chosen θ={best_th:.2f}")
ax.set_title("hate_speech Precision / Recall / F1 vs decision threshold")
ax.set_xlabel("θ  —  predict hate if P(hate) ≥ θ"); ax.legend()
plt.tight_layout(); plt.show()
```

**Feature importance interpretation:**

The top features pushing toward *hate_speech* are the specific slurs and targeted group-attack
phrases. The features pushing *against* hate_speech (toward offensive or neither) are generic
profanity that appears heavily in ordinary offensive conversation without a group target.

The model has learned the right associations. The residual errors — hate tweets classified as
offensive — occur when a tweet uses hateful vocabulary but the annotators judged it as ordinary
offensive language, or when the specific combination of tokens has not been seen in training.

**Threshold tradeoff:**

Lowering θ shifts recall up and precision down. The optimal θ (chosen on OOF predictions)
is where hate F1 peaks. Going below it raises recall further but false positives accumulate
faster, and hate F1 falls again. The chosen operating point balances both.

### The annotator confidence signal — two ways to use it

The confidence column (max_votes / total_votes) encodes label reliability directly.
Two complementary uses were explored:

1. **Dataset cleaning** — remove offensive rows with confidence < 0.67. These are the tweets
   sitting on the hate/offensive boundary in human perception. Removing them gives the model
   a cleaner training signal with less label noise near the decision boundary.

2. **Sample weighting** — pass confidence as `sample_weight` to the classifier. Every remaining
   tweet pulls the decision boundary in proportion to how certain its label is. Used together
   these two techniques are the most principled approach available without external data.

---

### Why there is a ceiling at hate F1 ≈ 0.46

Even with the best combination of the above, hate F1 plateaus around 0.46. Three factors set
this limit:

1. **Vocabulary overlap** — hate and offensive use the same slurs. Intent and target — the
   actual discriminators — are not visible in a bag of words.
2. **Residual label noise** — even after filtering, the retained hate labels carry only 73%
   mean agreement. Some are genuinely ambiguous even for humans.
3. **Only 915 hate training examples** — too few to learn the full diversity of hate speech.

Fine-tuned transformer models (DistilBERT, RoBERTa) are the natural next step: they encode
each token in the context of its full sentence via attention, making it possible to represent
who is addressing whom — the contextual signal that separates hate from offensive at the
semantic level. Published results on this dataset reach hate F1 ≈ 0.60–0.68 with BERT.

### Why there is a ceiling at hate F1 ≈ 0.45

Three factors set an upper bound that no feature-engineering approach can cross:

1. **Vocabulary overlap** — hate and offensive share the same slurs. The discriminative signal
   is *intent* and *target*, not which words appear. Surface features cannot encode this.
2. **Label noise** — annotators agreed only 73% of the time on hate labels. Some "errors" are
   genuine disagreements between humans, not model mistakes.
3. **Data scarcity** — only 915 hate tweets in the training set.

The natural next step is **fine-tuned transformer models** (DistilBERT, RoBERTa). They read
the full tweet as a sequence with attention, encoding who is addressing whom — the contextual
information that breaks the hate/offensive ambiguity. Published benchmarks on this dataset
report hate F1 ≈ 0.60–0.68 with BERT, roughly 15 points above the TF-IDF ceiling.

## 15. Does annotator COUNT add information beyond confidence?

The `confidence = majority_votes / count` signal used so far is a *ratio* — it ignores
**how many** annotators produced it. But `count` carries real information:

- `count` is never 1 in this dataset — the minimum is **3 annotators** (92% of tweets have
  exactly 3; the rest have 4, 6, 7 or 9).
- A unanimous **3/3** and a unanimous **9/9** both give `confidence = 1.0`, yet the 9/9 label is
  far more trustworthy. The ratio alone cannot see this.

So the question: can we use `count` to weight the data *better* than confidence alone?
Below, four weighting schemes are compared on the **best pipeline so far**
(TF-IDF + calibrated LinearSVC + OOF-tuned hate threshold), against the old confidence-only best.


### The four schemes

| Scheme | Formula | Idea |
|:--|:--|:--|
| **A — confidence only** (old best) | `majority / count` | baseline — ignores count |
| **B — confidence × count** | `confidence · count` | naive: more annotators ⇒ more weight |
| **C — confidence × log(count)** | `confidence · log₂(count)` | naive but dampened scaling |
| **D — Wilson lower bound** | lower bound of the agreement proportion at n = count | treats count as *uncertainty*: small n is penalised |
| **E — Bayesian shrinkage** | `(majority + k·p₀) / (count + k)` | pulls small-n agreement toward the global prior |

The key hypothesis: **count should be used as a measure of *uncertainty* (how sure we are of the
ratio), not as a raw multiplier.** Schemes D and E encode that; B and C are the naive baselines.
A unanimous 3/3 has a wider uncertainty interval than a unanimous 9/9 — Wilson and Bayesian
shrinkage both push the 3/3 weight down accordingly, which is exactly the effect we want.


```python
# ── Build the count-aware weight vectors from the SAME training split ──────────
# Reuses: df_train, X_tr_tfidf, X_te_tfidf, y_train, y_test, oof_calibrated, record, plot_cm

maj_tr = df_train[["hate_speech", "offensive_language", "neither"]].max(axis=1).values.astype(float)
n_tr   = df_train["count"].values.astype(float)
conf_tr = df_train["confidence"].values            # = maj_tr / n_tr  (Scheme A)

# B, C — naive scaling
w_B = conf_tr * n_tr
w_C = conf_tr * np.log2(n_tr)

# D — Wilson score lower bound of the majority proportion (z = 1.96)
z = 1.96
phat = maj_tr / n_tr
w_D = (phat + z*z/(2*n_tr) - z*np.sqrt((phat*(1-phat) + z*z/(4*n_tr)) / n_tr)) / (1 + z*z/n_tr)

# E — Bayesian shrinkage toward the global agreement prior p0, strength k
p0 = phat.mean(); k = 3.0
w_E = (maj_tr + k*p0) / (n_tr + k)

schemes = {
    "A — confidence only (old best)": conf_tr,
    "B — confidence x count"        : w_B,
    "C — confidence x log(count)"   : w_C,
    "D — Wilson lower bound"        : w_D,
    "E — Bayesian shrinkage"        : w_E,
}

print("Weight scheme summary (training set):")
for name, w in schemes.items():
    print(f"  {name:<32}  mean={w.mean():.3f}  min={w.min():.3f}  max={w.max():.3f}")
```

```python
# ── Evaluate each weight scheme on the best pipeline ───────────────────────────
# Pipeline = calibrated LinearSVC on TF-IDF, hate threshold tuned on OOF (never on test)

def eval_weight(sw, tag):
    sw = np.asarray(sw, dtype=float)
    # 1) OOF probabilities on the training set to tune theta
    oof = oof_calibrated(X_tr_tfidf, y_train.values, sw_e=sw, C=0.1)
    best_th, best_h = 0.5, 0.0
    for th in np.arange(0.05, 0.50, 0.01):
        p = np.where(oof[:, 0] >= th, 0, np.argmax(oof[:, 1:], axis=1) + 1)
        h = f1_score(y_train.values, p, labels=[0], average="micro", zero_division=0)
        if h > best_h:
            best_h, best_th = h, th
    # 2) fit on full train, predict test ONCE
    cal = CalibratedClassifierCV(
        LinearSVC(C=0.1, class_weight="balanced", random_state=SEED), cv=5)
    cal.fit(X_tr_tfidf, y_train, sample_weight=sw)
    tp = cal.predict_proba(X_te_tfidf)
    yp = np.where(tp[:, 0] >= best_th, 0, np.argmax(tp[:, 1:], axis=1) + 1)
    record("CountWeight", f"{tag} (theta={best_th:.2f})", y_test.values, yp)
    return {
        "scheme": tag, "theta": round(best_th, 2),
        "hate_F1": round(f1_score(y_test, yp, labels=[0], average="micro", zero_division=0), 3),
        "hate_P":  round(precision_score(y_test, yp, labels=[0], average="micro", zero_division=0), 3),
        "hate_R":  round(recall_score(y_test, yp, labels=[0], average="micro", zero_division=0), 3),
        "Macro_F1": round(f1_score(y_test, yp, average="macro"), 3),
        "Accuracy": round(accuracy_score(y_test, yp), 3),
    }, yp

rows, preds = [], {}
for tag, sw in schemes.items():
    r, yp = eval_weight(sw, tag)
    rows.append(r); preds[tag] = yp

count_df = pd.DataFrame(rows)
print(count_df.to_string(index=False))
```

```python
# ── Visualise: count-aware schemes vs confidence-only baseline ─────────────────
base = count_df.iloc[0]   # Scheme A

fig, axes = plt.subplots(1, 2, figsize=(14, 4.5))
metrics = ["hate_F1", "hate_R", "Macro_F1"]
x = np.arange(len(count_df)); w = 0.25
colors = ["#94a3b8", "#d62728", "#5b21b6"]
for i, m in enumerate(metrics):
    axes[0].bar(x + i*w, count_df[m], w, label=m, color=colors[i], alpha=0.85)
axes[0].axhline(base["hate_F1"], color="#d62728", ls=":", lw=1, alpha=0.6)
axes[0].set_xticks(x + w); axes[0].set_xticklabels(
    [s.split(" — ")[0] for s in count_df["scheme"]], rotation=0)
axes[0].set_title("Count-weighting schemes — key metrics"); axes[0].legend(fontsize=8)
axes[0].set_ylim(0, 0.85)

# confusion matrix of the best count-aware scheme (highest hate_F1)
best_tag = count_df.sort_values(["hate_F1", "hate_R"], ascending=False).iloc[0]["scheme"]
cm = confusion_matrix(y_test, preds[best_tag])
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="Purples",
            xticklabels=TARGET_NAMES, yticklabels=TARGET_NAMES, ax=axes[1])
axes[1].set_title(f"Best count-aware scheme (row-norm = recall)\n{best_tag}")
axes[1].set_xlabel("Predicted"); axes[1].set_ylabel("True")
plt.tight_layout(); plt.show()

print(f"Best count-aware scheme: {best_tag}")
```

### Verdict — does `count` help?

**Yes, but only when used correctly.** The naive schemes (B, C) that multiply confidence by
count actually **hurt** performance. The reason is structural: 92% of tweets have exactly
`count = 3`, so multiplying mostly rescales every weight identically and lets the handful of
6–9-annotator tweets dominate the loss — that is noise, not signal.

The principled schemes treat `count` as **uncertainty**, not magnitude:

- **Wilson lower bound (D)** gives the best **hate F1 and hate recall** — it penalises a unanimous
  3/3 more than a unanimous 9/9, exactly the distinction the plain ratio missed. The cost is a
  negligible drop in macro F1.
- **Bayesian shrinkage (E)** trades recall for the **highest precision and accuracy** — useful if
  false positives are more costly than missed hate.

**Conclusion:** annotator `count` does carry information beyond the confidence ratio, but it must
be folded in as a confidence-interval / shrinkage adjustment (D or E), never as a raw multiplier
(B or C). For maximising hate detection, the **Wilson lower bound weight** is the new best
configuration; for a precision-leaning operating point, **Bayesian shrinkage** wins.
