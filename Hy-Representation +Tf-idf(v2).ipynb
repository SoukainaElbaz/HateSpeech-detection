 # 1. text cleaning

```python
import re
import emoji

# Known hate / toxic codes as whole tokens; longest first so masking is unambiguous
hate_codes_strong = {"1488", "88", "14", "18", "28"}
hate_codes_medium = {"311", "33", "333"}
toxic_codes = {"1312", "13", "187"}
weak_codes = {"666"}
extra_preserved_codes = {"69", "911", "12", "12.666"}

PRESERVED_NUMBER_CODES = sorted(
    hate_codes_strong
    | hate_codes_medium
    | toxic_codes
    | weak_codes
    | extra_preserved_codes,
    key=len,
    reverse=True,
)
_PH_PREFIX = "__PRES_NUM_"
PRESERVED_NUMBER_CODES_SET = set(PRESERVED_NUMBER_CODES)


def _protect_preserved_number_codes(text):
    # Replace each whitelisted number (whole word) with a placeholder we restore later
    for i, code in enumerate(PRESERVED_NUMBER_CODES):
        ph = f" {_PH_PREFIX}{i}__ "
        pat = r"\b" + re.escape(code) + r"\b"
        text = re.sub(pat, ph, text)
    return text


def _restore_preserved_number_codes(text):
    # Turn placeholders back into the original digit strings
    for i, code in enumerate(PRESERVED_NUMBER_CODES):
        ph = f" {_PH_PREFIX}{i}__ "
        text = text.replace(ph, f" {code} ")
    return text


def _strip_standalone_numbers(text):
    # Drop integers/decimals that form their own "word" (digits inside b1tch stay, no \b on the middle 1)
    return re.sub(r"\b\d+(?:\.\d+)?\b", " ", text)


def process_token(token):
    # After cleaning, a numeric-only token should be a preserved code, or be dropped
    if re.fullmatch(r"\d+(?:\.\d+)?", token):
        if token in PRESERVED_NUMBER_CODES_SET:
            return token
        return ""
    return token


def clean_text(text, lang="en"):
    # lang: unused after switching from num2words; kept for a stable function signature
    if not isinstance(text, str):
        text = str(text)  # Coerce to string so the rest of the pipeline is safe
    if not text:
        return text  # Nothing to do for empty / falsy text

    text = emoji.demojize(text)  # :smile: text instead of pictographs, before we strip symbols
    text = text.lower()  # Make casing uniform for GloVe / matching

    text = re.sub(r"@\w+", " ", text)  # Remove @username mentions
    text = re.sub(r"\brt\b", " ", text)  # Remove retweet prefix as its own word

    text = _protect_preserved_number_codes(text)  # So we do not delete allowed codes next
    text = _strip_standalone_numbers(text)  # Remove other stand-alone numbers

    text = text.replace("-", " ")  # Turn hyphens (e.g. forty-two) into separable word chunks
    text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation; keep word chars and whitespace

    text = _restore_preserved_number_codes(text)  # After punct so decimals like 12.666 keep the dot

    tokens = text.split()  # Whitespace tokenization
    tokens = [process_token(t) for t in tokens if t]  # drop stray number tokens
    text = " ".join(tokens)  # Single string again for later steps
    text = re.sub(r"\s+", " ", text).strip()  # One space between words, trim ends
    return text
```

# 2. tokenization

```python
def tokenize_text(text):
    tokens = text.split()
    return tokens
```

```python
clean_comment = clean_text("You are STUPID!!! 😡 123")
tokenize_text(clean_comment)
```

# 3. stopword

```python
nltk.download('stopwords', quiet=True)

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def remove_stopwords(tokens):
    return [w for w in tokens if w not in stop_words]
```

```python
tokens = ["you", "are", "a", "stupid", "person"]
remove_stopwords(tokens)
```

# 4. lemmatization

```python
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(tokens):
    return [lemmatizer.lemmatize(w, pos='v') for w in tokens]
```

```python
tokens = ["running", "hated", "better"]
lemmatize_text(tokens)
```

```python
def full_preprocess(text):
    if pd.isna(text):
        return ""
    
    text = clean_text(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize_text(tokens)
    
    return " ".join(tokens)
```

```python
df["cleaned_text"] = df["tweet"].apply(full_preprocess)
```

# 5. vectorization (Embedding)

## word2vect

```python
sentences = [
    row.split()
    for row in df["cleaned_text"].astype(str)
    if row.strip()
]
# par defaut cbow, sg=0
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=2, 
    epochs=5,
)
```

```python
word = "trash"

if word in model.wv:
    vector = model.wv[word]
    print(vector)
else:
    print(word, "not in vocabulary")
```

## FastText

```python
sentences = [
    row.split()
    for row in df["cleaned_text"].astype(str)
    if row.strip()
]
fasttext_model = FastText(
    sentences,
    vector_size=100,
    window=5,
    min_count=2,
    epochs=5
)
```

```python
word = "trash"
print("FastText:", fasttext_model.wv[word][:5])
```

## GloVe

```python
# ===============================
# GloVe — Load Pretrained Vectors
# ===============================

glove_path = "glove.6B.100d.txt"

glove_embeddings = {}

with open(glove_path, encoding="utf8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        glove_embeddings[word] = vector

print("Loaded", len(glove_embeddings), "GloVe word vectors")
```

```python
# ===============================
# Test Word Vector
# ===============================

word = "trash"

if word in glove_embeddings:
    vector = glove_embeddings[word]
    print("GloVe:", vector[:5])  # show first 5 numbers
else:
    print(word, "not in GloVe vocabulary")
```
